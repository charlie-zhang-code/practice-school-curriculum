根据“post_data”中views列的数据量将category列生成词云图。

```python
import chardet # type: ignore
import pandas as pd

with open('post_data.csv', 'rb') as f:
    result = chardet.detect(f.read())
    encoding = result['encoding']
    print("文件编码：" + result['encoding'])

pr = pd.DataFrame(pd.read_csv('post_data.csv', sep = ',', header=0, encoding=encoding))
pr
```

```python
# 计算每个category的总views
category_views = pr.groupby('category')['views'].sum()
category_views
```

```python
# 将views转换为字符串，并作为权重
weights = {category: views for category, views in category_views.items()}
weights
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

font_path = 'C:/Windows/Fonts/simsun.ttc'
# 生成词云图
wordcloud = WordCloud(
    width=800,
    height=250,
    background_color='white',
    font_path=font_path
).generate_from_frequencies(weights)

# 显示词云图
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

利用jieba库对“text”进行分词，并生成词云图。

```python
import chardet # type: ignore

with open('text.txt', 'rb') as f:
    result = chardet.detect(f.read())
    encoding = result['encoding']
    print("文件编码：" + result['encoding'])
    text = f.read()

# 读取文本文件
with open('text.txt', 'r', encoding=encoding) as file:
    text = file.read()

text
```

```python
import jieba

# 使用jieba进行分词
words = jieba.cut(text)
word_list = ' '.join(words)

word_list
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

font_path = 'C:/Windows/Fonts/simsun.ttc'
# 生成词云图
wordcloud = WordCloud(
    width=800,
    height=250,
    background_color='white',
    font_path=font_path
).generate(word_list)

# 显示词云图
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

利用“weibo-2”json数据完成微博转发关系图的绘制。

```python
import chardet # type: ignore

with open('weibo-2.json', 'rb') as f:
    result = chardet.detect(f.read())
    encoding = result['encoding']
    print("文件编码：" + result['encoding'])

# 读取JSON文件
import json
with open('weibo-2.json', 'r', encoding=encoding) as file:
    weibo_json = json.load(file)

weibo_json
```

```python
from pyecharts import options as opts
from pyecharts.charts import Graph

graph = (
    Graph(init_opts=opts.InitOpts(width="100%", height="650px")) 
    .add(
        "",
        weibo_json[0], # 节点数据
        weibo_json[1], # 边数据
        weibo_json[2], # 分类数据
        repulsion=50,
        linestyle_opts=opts.LineStyleOpts(curve=0.2),
        label_opts=opts.LabelOpts(is_show=False),
    )
    .set_global_opts(
        legend_opts=opts.LegendOpts(is_show=False),
        title_opts=opts.TitleOpts(title="微博转发关系图"),
    )
)

graph.render("weibo_repost_graph.html")
```
