1.利用所给vote_result数据绘制饼图，饼图中添加标题为“用户感兴趣领域”，设置合适的字体大小，显示比例和各模块详细的文本说明，且图例垂直显示在饼图的右边。

```python
import pandas as pd
import chardet

with open('vote_result.csv', 'rb') as f:
    result = chardet.detect(f.read())

pr = pd.DataFrame(pd.read_csv('vote_result.csv', sep = ',', header=0, encoding=result['encoding']))
print("文件编码：" + result['encoding'])
pr
```

```python
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体
plt.rcParams['axes.unicode_minus'] = False  # 正确显示负号

interests = pr['感兴趣的领域']
votes = pr['票数']

# 绘制饼图
plt.figure(figsize=(8, 5))  # 设置图形的大小
plt.pie(votes, labels=interests, autopct='%1.1f%%', startangle=140)

plt.title('用户感兴趣领域', fontsize=16)
plt.legend(interests, title="领域", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
plt.show()
```

2.利用任务1中的数据，使用matplotlib完成环形图的绘制；

```python
plt.figure(figsize=(8, 5))  # 设置图形的大小
plt.pie(votes, labels=interests, autopct='%1.1f%%', startangle=140, wedgeprops=dict(width=0.3))

plt.title('用户感兴趣领域', fontsize=16)
plt.legend(interests, title="领域", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
plt.show()
```
