```
pr1=pd.DataFrame({"id":[1001,1002,1003,1000,10011,1006,10067,10088，1009],
```

将pr和pr1两个数据表合并，交集，并集，'left'，'right'。

````python
# 创建数据表pr1
pr1 = pd.DataFrame({
    "id": [1001, 1002, 1003, 1000, 10011, 1006, 10067, 10088, 1009],
    "gender": ['male', 'female', 'male', 'female', 'male', 'female', 'male', 'female', 'female'],
    "pay": ['Y', 'N', 'Y', 'Y', 'N', 'Y', 'N', 'Y', 'Y'],
    "m-point": [10, 12, 20, 40, 40, 40, 30, 20, 30]
})

pr1
````

````python
pr
````

````python
# 交集（inner join）
pr_inner_join_pr1 = pd.merge(pr, pr1, on='id', how='inner')
pr_inner_join_pr1
````

````python
# 并集（full outer join）
pr_full_outer_join_pr1 = pd.merge(pr, pr1, on='id', how='outer')
pr_full_outer_join_pr1
````

````python
# 左连接（left join）
pr_left_join_pr1 = pd.merge(pr, pr1, on='id', how='left')
pr_left_join_pr1
````

````python
# 右连接（right join）
pr_right_join_pr1 = pd.merge(pr, pr1, on='id', how='right')
pr_right_join_pr1
````

（8）使用与、或、非三个条件配合大于、小于、等于对数据进行筛选，并进行计数和求和。
使用“与”进行筛选。筛选出SQ_FT中数值大于1000且ST_NAME中值为BERKELEY所在的行。

````python
# 重新读取数据，上面的数据已经修改，没有SQ_FT列
pr = pd.DataFrame(pd.read_csv('property-data.csv', sep = ',', header=0))
pr['SQ_FT'] = pd.to_numeric(pr['SQ_FT'], errors='coerce')

# 条件筛选
filtered_rows = pr[(pr['SQ_FT'] > 1000) & (pr['ST_NAME'] == 'BERKELEY')]
filtered_rows
````

````python
# 计数
count_filtered = filtered_rows.shape[0]
count_filtered
````

````python
sum_filtered = filtered_rows['ST_NUM'].sum() 
sum_filtered
````

（9）使用“或”进行筛选,筛选出SQ_FT中数值大于1000所在的列或ST_NAME中值为BERKELEY 所在的行，并将SQ_FT中数据值降序排列。

````python
filtered_rows = pr[(pr['SQ_FT'] > 1000) | (pr['ST_NAME'] == 'BERKELEY')]
filtered_rows
````

````python
filtered_rows_sort = filtered_rows.sort_values(by='SQ_FT', ascending=False)
filtered_rows_sort
````

（10）读取新生数据文件数据，按性别分组计数。

````python
import pandas as pd
baby_data = pd.read_excel('新生数据.xls', index_col=0)
baby_data
````

````python
gender_counts = baby_data['性别'].value_counts()
gender_counts
````

（11）按性别字段进行汇总，并分别计算身高的合计和均值。

````python
# 按性别分组，并计算身高的合计和均值
gender_summary = baby_data.groupby('性别')['身高'].agg(['sum', 'mean']).reset_index()
gender_summary
````

（12）对读取的新生数据随机采样15个样本，并将采样后结果放回和不放回结果显示出来。

````python
# 不放回随机采样15个样本
sample_without_replacement = baby_data.sample(n=15)
sample_without_replacement
````

````python
# 放回随机采样15个样本
sample_with_replacement = baby_data.sample(n=15, replace=True)
sample_with_replacement
````
